Load the dataset and explore the variables.

```python
import pandas as pd

# Cargar el archivo
data = pd.read_csv(r'C:\Users\Win10\Downloads\customer_churn.csv')

# Explorar las primeras filas del dataset
data.head()
```

We will try to predict variable Churn using a logistic regression on variables tenure, SeniorCitizen,MonthlyCharges.

Extract the target variable.

```python
# Extraer la variable objetivo
objetivo = data['Churn']
```

Extract the independent variables and scale them.

```python
from sklearn.preprocessing import StandardScaler

# Extraer las variables independientes
variables_independientes = data[['tenure', 'SeniorCitizen', 'MonthlyCharges']]

# Escalar las variables
escalador = StandardScaler()
variables_escaladas = escalador.fit_transform(variables_independientes)
```

Build the logistic regression model.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(variables_escaladas, objetivo, test_size=0.2, random_state=42)

# Crear el modelo de regresión logística
modelo = LogisticRegression()

# Entrenar el modelo
modelo.fit(X_train, y_train)
```

Evaluate the model.

```python
from sklearn.metrics import accuracy_score

# Realizar predicciones en el conjunto de prueba
predicciones = modelo.predict(X_test)

# Calcular la precisión
precision = accuracy_score(y_test, predicciones)
print(f'Precisión del modelo: {precision}')
```

Even a simple model will give us more than 70% accuracy. Why?

un modelo simple puede dar más del 70% de precisión por el desbalance de clases en el dataset. Si la mayoría de los clientes no se dieron de baja, el modelo puede predecir la clase mayoritaria con alta precisión sin ser realmente efectivo para identificar los casos de baja.

Synthetic Minority Oversampling TEchnique (SMOTE) is an over sampling technique based on nearest neighbors that adds new points between existing points. Apply imblearn.over_sampling.SMOTE to the dataset. Build and evaluate the logistic regression model. Is it there any improvement?

```python
from imblearn.over_sampling import SMOTE

# Aplicar SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(variables_escaladas, objetivo)

# Dividir los datos balanceados en conjunto de entrenamiento y prueba
X_train_res, X_test_res, y_train_res, y_test_res = train_test_split(X_res, y_res, test_size=0.2, random_state=42)

# Entrenar el modelo con los datos balanceados
modelo_res = LogisticRegression()
modelo_res.fit(X_train_res, y_train_res)

# Realizar predicciones en el conjunto de prueba
predicciones_res = modelo_res.predict(X_test_res)

# Calcular la precisión
precision_res = accuracy_score(y_test_res, predicciones_res)
print(f'Precisión del modelo con SMOTE: {precision_res}')
```

Tomek links are pairs of very close instances, but of opposite classes. Removing the instances of the majority class of each pair increases the space between the two classes, facilitating the classification process. Apply imblearn.under_sampling.TomekLinks to the dataset. Build and evaluate the logistic regression model. Is it there any improvement?

```python
from imblearn.under_sampling import TomekLinks

# Aplicar Tomek Links
tomek = TomekLinks()
X_res_tomek, y_res_tomek = tomek.fit_resample(variables_escaladas, objetivo)

# Dividir los datos reducidos en conjunto de entrenamiento y prueba
X_train_tomek, X_test_tomek, y_train_tomek, y_test_tomek = train_test_split(X_res_tomek, y_res_tomek, test_size=0.2, random_state=42)

# Entrenar el modelo con los datos reducidos
modelo_tomek = LogisticRegression()
modelo_tomek.fit(X_train_tomek, y_train_tomek)

# Realizar predicciones en el conjunto de prueba
predicciones_tomek = modelo_tomek.predict(X_test_tomek)

# Calcular la precisión
precision_tomek = accuracy_score(y_test_tomek, predicciones_tomek)
print(f'Precisión del modelo con Tomek Links: {precision_tomek}')
```
